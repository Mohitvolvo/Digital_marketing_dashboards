```python
import random
import pandas as pd
from faker import Faker


fake = Faker()

# Constants
NUM_RECORDS = 1000
PLATFORMS = ['Twitter', 'Facebook', 'Instagram', 'LinkedIn']
SENTIMENTS = ['Positive', 'Neutral', 'Negative']
KEYWORDS_HASHTAGS = ['#awesome', '#great', '#good', '#okay', '#bad', '#terrible', 'product', 'service', 'experience', 'feedback']

# Generate fake social listening data
social_data = []
for _ in range(NUM_RECORDS):
    date = fake.date_this_year()
    platform = random.choice(PLATFORMS)
    sentiment = random.choice(SENTIMENTS)
    keyword = random.choice(KEYWORDS_HASHTAGS)
    share_of_voice = random.randint(1, 100)
    influencer = fake.name()
    engagement = random.randint(10, 1000)
    
    social_data.append({
        'date': date,
        'platform': platform,
        'sentiment': sentiment,
        'keyword': keyword,
        'share_of_voice': share_of_voice,
        'influencer': influencer,
        'engagement': engagement
    })

social_df = pd.DataFrame(social_data)

# Export data to CSV
social_df.to_csv('social_data.csv', index=False)
```

```python
import random
import pandas as pd
from faker import Faker

fake = Faker()

# Constants
NUM_RECORDS = 1000
PLATFORMS = ['Twitter', 'Facebook', 'Instagram', 'LinkedIn', 'YouTube']
SENTIMENTS = ['Positive', 'Neutral', 'Negative']
INFLUENCERS = ['Influencer1', 'Influencer2', 'Influencer3', 'Influencer4', 'Influencer5']

# Generate fake sentiment data
data = []
for _ in range(NUM_RECORDS):
    date = fake.date_this_year()
    platform = random.choice(PLATFORMS)
    sentiment = random.choice(SENTIMENTS)
    mentions = random.randint(1, 500)
    engagement = random.randint(100, 10000)
    influencer = random.choice(INFLUENCERS)
    keywords = fake.words(nb=random.randint(1, 5), unique=True)
    keywords_str = ' '.join(keywords)
    
    data.append({
        'date': date,
        'platform': platform,
        'sentiment': sentiment,
        'mentions': mentions,
        'engagement': engagement,
        'influencer': influencer,
        'keywords': keywords_str
    })

df = pd.DataFrame(data)

# Export data to CSV
df.to_csv('sentiment_data.csv', index=False)
```
