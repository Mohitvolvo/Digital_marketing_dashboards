```python
import random
import pandas as pd
from faker import Faker


fake = Faker()

# Constants
NUM_CAMPAIGNS = 1000
PLATFORMS = ['Google Ads', 'Facebook Ads', 'Instagram Ads', 'LinkedIn Ads']

# Generate fake campaign data
campaign_data = []
for _ in range(NUM_CAMPAIGNS):
    start_date = fake.date_this_year()
    end_date = fake.date_between(start_date=start_date, end_date='today')
    platform = random.choice(PLATFORMS)
    impressions = random.randint(1000, 100000)
    clicks = random.randint(100, 10000)
    spend = round(random.uniform(500, 50000), 2)
    conversions = random.randint(10, 1000)
    cpc = spend / clicks
    conversion_rate = conversions / clicks
    revenue = round(spend * random.uniform(1.2, 2.0), 2)
    roas = revenue / spend
    cpa = spend / conversions

    campaign_data.append({
        'start_date': start_date,
        'end_date': end_date,
        'platform': platform,
        'impressions': impressions,
        'clicks': clicks,
        'spend': spend,
        'conversions': conversions,
        'cpc': cpc,
        'conversion_rate': conversion_rate,
        'revenue': revenue,
        'roas': roas,
        'cpa': cpa
    })

campaign_df = pd.DataFrame(campaign_data)

# Export data to CSV
campaign_df.to_csv('campaign_data.csv', index=False)
```

```python
import random
import pandas as pd
from faker import Faker

fake = Faker()

# Constants
NUM_RECORDS = 1000
CAMPAIGNS = ['Campaign1', 'Campaign2', 'Campaign3', 'Campaign4', 'Campaign5']
PLATFORMS = ['Google Ads', 'Facebook Ads', 'Instagram Ads', 'LinkedIn Ads', 'Twitter Ads']

# Generate fake advertising data
ad_data = []
for _ in range(NUM_RECORDS):
    date = fake.date_this_year()
    campaign = random.choice(CAMPAIGNS)
    platform = random.choice(PLATFORMS)
    impressions = random.randint(1000, 100000)
    clicks = random.randint(100, 10000)
    cost = round(random.uniform(100, 10000), 2)
    conversions = random.randint(10, 1000)
    cpc = round(cost / clicks, 2) if clicks > 0 else 0
    conversion_rate = round(conversions / clicks, 4) if clicks > 0 else 0
    roas = round(random.uniform(1, 10), 2)
    cpa = round(cost / conversions, 2) if conversions > 0 else 0

    ad_data.append({
        'date': date,
        'campaign': campaign,
        'platform': platform,
        'impressions': impressions,
        'clicks': clicks,
        'cost': cost,
        'conversions': conversions,
        'cpc': cpc,
        'conversion_rate': conversion_rate,
        'roas': roas,
        'cpa': cpa
    })

ad_df = pd.DataFrame(ad_data)

# Export data to CSV
ad_df.to_csv('ad_data.csv', index=False)
```
