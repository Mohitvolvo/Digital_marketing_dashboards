```python
import random
import pandas as pd
from faker import Faker


fake = Faker()

# Constants
NUM_POSTS = 1000
NUM_DAYS = 365
PLATFORMS = ['Facebook', 'Twitter', 'Instagram', 'LinkedIn']
CONTENT_TYPES = ['video', 'image', 'text']

# Generate fake posts data
posts_data = []
for _ in range(NUM_POSTS):
    post_date = fake.date_time_this_year()
    platform = random.choice(PLATFORMS)
    content_type = random.choice(CONTENT_TYPES)
    likes = random.randint(0, 1000)
    shares = random.randint(0, 500)
    comments = random.randint(0, 300)
    impressions = random.randint(1000, 10000)
    engagements = likes + shares + comments
    engagement_rate = engagements / impressions

    posts_data.append({
        'date': post_date,
        'platform': platform,
        'content_type': content_type,
        'likes': likes,
        'shares': shares,
        'comments': comments,
        'impressions': impressions,
        'engagements': engagements,
        'engagement_rate': engagement_rate
    })

posts_df = pd.DataFrame(posts_data)

# Generate fake follower growth data
follower_growth_data = []
for day in range(NUM_DAYS):
    date = fake.date_this_year()
    new_followers = random.randint(0, 100)
    follower_growth_data.append({
        'date': date,
        'new_followers': new_followers
    })

follower_growth_df = pd.DataFrame(follower_growth_data)
follower_growth_df['date'] = pd.to_datetime(follower_growth_df['date'])
follower_growth_df = follower_growth_df.sort_values('date').reset_index(drop=True)
follower_growth_df['total_followers'] = follower_growth_df['new_followers'].cumsum()

# Export data to CSV
posts_df.to_csv('posts_data.csv', index=False)
follower_growth_df.to_csv('follower_growth_data.csv', index=False)



```
