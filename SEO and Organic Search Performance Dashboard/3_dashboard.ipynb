```python
import random
import pandas as pd
from faker import Faker

fake = Faker()

# Constants
NUM_RECORDS = 1000
NUM_DAYS = 365
KEYWORDS = ['keyword1', 'keyword2', 'keyword3', 'keyword4', 'keyword5']

# Generate fake SEO data
seo_data = []
for _ in range(NUM_RECORDS):
    date = fake.date_this_year()
    keyword = random.choice(KEYWORDS)
    organic_traffic = random.randint(100, 10000)
    keyword_ranking = random.randint(1, 100)
    ctr = round(random.uniform(0.01, 0.5), 4)
    bounce_rate = round(random.uniform(0.1, 1.0), 2)
    avg_session_duration = round(random.uniform(30, 600), 2)
    backlinks = random.randint(0, 500)
    domain_authority = random.randint(1, 100)
    
    seo_data.append({
        'date': date,
        'keyword': keyword,
        'organic_traffic': organic_traffic,
        'keyword_ranking': keyword_ranking,
        'ctr': ctr,
        'bounce_rate': bounce_rate,
        'avg_session_duration': avg_session_duration,
        'backlinks': backlinks,
        'domain_authority': domain_authority
    })

seo_df = pd.DataFrame(seo_data)

# Export data to CSV
seo_df.to_csv('seo_data.csv', index=False)
```

```python
import random
import pandas as pd
from faker import Faker
import matplotlib.pyplot as plt
import seaborn as sns

fake = Faker()

# Constants
NUM_RECORDS = 1000
NUM_DAYS = 365
KEYWORDS = ['keyword1', 'keyword2', 'keyword3', 'keyword4', 'keyword5']
SOURCES_MEDIUMS = ['Organic Search', 'Direct', 'Referral', 'Social']

# Generate fake SEO data
seo_data = []
for _ in range(NUM_RECORDS):
    date = fake.date_this_year()
    keyword = random.choice(KEYWORDS)
    source_medium = random.choice(SOURCES_MEDIUMS)
    organic_traffic = random.randint(100, 10000)
    keyword_ranking = random.randint(1, 100)
    ctr = round(random.uniform(0.01, 0.5), 4)
    bounce_rate = round(random.uniform(0.1, 1.0), 2)
    avg_session_duration = round(random.uniform(30, 600), 2)
    backlinks = random.randint(0, 500)
    domain_authority = random.randint(1, 100)
    
    seo_data.append({
        'date': date,
        'keyword': keyword,
        'source_medium': source_medium,
        'organic_traffic': organic_traffic,
        'keyword_ranking': keyword_ranking,
        'ctr': ctr,
        'bounce_rate': bounce_rate,
        'avg_session_duration': avg_session_duration,
        'backlinks': backlinks,
        'domain_authority': domain_authority
    })

seo_df = pd.DataFrame(seo_data)

# Export data to CSV
seo_df.to_csv('seo_data_v1.csv', index=False)
```
